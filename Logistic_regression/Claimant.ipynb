# Import Necessary Libraries

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

# Import Data

```python
claimants_data=pd.read_csv("Claimant.csv")
claimants_data
```

# 4. Data Preparation

## 4.1 Data Cleaning

```python
claimants_data.dropna(inplace= True) #istead of giving new variable name we can use inplace = true which directly impact on the orginal data
```

```python
claimants_data.isna().sum()
```

```python
claimants_data.shape
```

## 4.2 Seperate input and output

```python
X =claimants_data.drop(labels=["CASENUM","ATTORNEY"],axis=1 )
X
```

```python
y= claimants_data["ATTORNEY"]
y
```

# 5. Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
logistic_model=LogisticRegression()
logistic_model
```

# 6. Model Training

```python
logistic_model.fit(X,y)
```

```python
logistic_model.get_params()
```

# 7. Model Testing

```python
y_predict=logistic_model.predict(X) #return the output (y)
y_predict
```

# 8. Model Evaluation 

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,y_predict)
```

# 9. Model Deployment

```python
from pickle import dump
```

```python
dump(logistic_model,open("Claimant.pkl","wb"))
```
